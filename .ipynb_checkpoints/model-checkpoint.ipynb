# STEP 0: Libraries Import Karna


```python
# Data handling ke liye
import pandas as pd
import numpy as np

# Visualization ke liye
import matplotlib.pyplot as plt

# ML ke liye
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# STEP 1: Thoda Ajeeb / Dirty Dataset Banana


```python
# Yahan hum jaan-bujh ke ganda data bana rahe hain
data = {
    'hours_study': [1, 2, 3, np.nan, 5, 6, 7, 20],   # 20 = outlier, NaN = missing
    'attendance': [50, 55, None, 65, 70, 300, 80, 85],  # 300 galat hai
    'prev_score': [40, 42, 45, 50, -10, 65, 70, 80],   # -10 invalid
    'pass': [0, 0, 0, 0, 1, 1, 1, 1]
}

df = pd.DataFrame(data)
df
```

# STEP 2: Data Understanding (EDA – Part 1)

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['hours_study'] = df['hours_study'].fillna(df['hours_study'].mean())
df['attendance'] = df['attendance'].fillna(df['attendance'].median())
```

```python
median_score = int(df['prev_score'].median())
df.loc[df['prev_score'] < 0, 'prev_score'] = median_score
```

```python
df
```

```python
plt.scatter(df['hours_study'], df['pass'])
plt.xlabel("Hours Studied")
plt.ylabel("Pass (0 = Fail, 1 = Pass)")
plt.show()
```

# STEP 5: Feature Engineering


```python
# Combined effort ka naya feature
df['study_effort'] = df['hours_study'] * df['attendance']
df
```

# STEP 6: Feature Selection

```python
# Input features
X = df[['hours_study', 'attendance', 'prev_score', 'study_effort']]

# Target variable
y = df['pass']
```

# STEP 7: Train–Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,     # 25% test data
    random_state=42
)
```

# STEP 8: Model Banana (Logistic Regression)

```python
from sklearn.preprocessing import StandardScaler

# Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Model create & train
model = LogisticRegression(solver='liblinear', random_state=42)
model.fit(X_train_scaled, y_train)

```

# STEP 9: Prediction

```python
y_pred = model.predict(X_test_scaled)
y_pred
```

# STEP 10: Model Evaluation

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

# Classification Report
cr = classification_report(y_test, y_pred)
print("Classification Report:\n", cr)
```

# Step 11: Model Accuracy

```python
# Accuracy in percentage
accuracy_percent = accuracy * 100
print("Model Accuracy: {:.2f}%".format(accuracy_percent))
```

# STEP 12: Predict for a new student

```python
# New student raw data
new_student = pd.DataFrame({
    'hours_study': [4],
    'attendance': [72],
    'prev_score': [55]
})

# Feature engineering (same as training)
new_student['study_effort'] = new_student['hours_study'] * new_student['attendance']

# Scaling (same scaler)
new_student_scaled = scaler.transform(new_student)

# Prediction
new_prediction = model.predict(new_student_scaled)
print("Prediction for new student:", new_prediction)
```
